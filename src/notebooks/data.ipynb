```python

raw_data_dir = Path('../../data/raw')
save_dir = Path('../../data/raw/combined')


def load_combined_parquet(parquet_dir='../../data/raw/combined'):
    """
    Load combined users and trips data from parquet files.
    
    Args:
        parquet_dir (str): Directory containing the parquet files
        
    Returns:
        tuple: (users_df, trips_df) DataFrames
    """
    parquet_path = Path(parquet_dir)
    
    users_file = parquet_path / 'users.parquet'
    trips_file = parquet_path / 'trips.parquet'
    
    if not users_file.exists() or not trips_file.exists():
        raise FileNotFoundError(f"Parquet files not found in {parquet_path}")
    
    users_df = pd.read_parquet(users_file)
    trips_df = pd.read_parquet(trips_file)
    
    return users_df, trips_df


users_df, trips_df = load_combined_parquet()
print("Loaded data from parquet files")

# print("Parquet files not found, loading from raw data...")
# users_df, trips_df = load_all_raw_data(raw_data_dir, save_dir, verbose=True, load_parquet=True)


print(f"Datos cargados: {len(users_df):,} usuarios, {len(trips_df):,} viajes")
```

```python

users_train = data_splits['users_train']
users_val = data_splits['users_val'] 
users_test = data_splits['users_test']

trips_train = data_splits['trips_train']
trips_val = data_splits['trips_val']
trips_test = data_splits['trips_test']


print(f"Users: Train={len(users_train):,}, Val={len(users_val):,}, Test={len(users_test):,}")
print(f"Trips: Train={len(trips_train):,}, Val={len(trips_val):,}, Test={len(trips_test):,}")
print(f"Final columns: {list(trips_train.columns)}")
```
