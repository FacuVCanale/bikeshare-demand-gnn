```python
df_merged = merge_intra_clusters(pdf_clustered)
```

```python
# Apply targeted outlier removal per cluster for trip_dur_mean_last_DT
df_merged = remove_top_outliers_per_cluster(df_merged, 'trip_dur_mean_last_DT', top_n=5)
```

```python
feature_cols = [
    'dep_last_DT', # "station_id",   # removed for generalization
    'trip_dur_mean_last_DT',
    'dep_lag_1',
    'dep_lag_2',
    'dep_lag_3',
    'dep_lag_4',
    'dep_lag_5',
    'dep_lag_6',
    'arr_last_DT',
    'arr_lag_1',
    'arr_lag_2',
    'arr_lag_3',
    'arr_lag_4',
    'arr_lag_5',
    'arr_lag_6',
    'weather_temperature_2m',
    'weather_relative_humidity_2m',
    'weather_apparent_temperature',
    'weather_precipitation',
    'weather_wind_speed_10m',
    #'weather_sunshine_duration',
    'weather_is_day',
    #'lat_mean',
    #'lon_mean',
    #'station_name',
    #'hour',
    #'dow',
    #'month',
    #'day',
    'is_holiday_ar',
    'sin_hour',
    'cos_hour',
    'sin_dow',
    'cos_dow',
    'sin_month',
    'cos_month',
    'is_weekend',
    'payday_flag',
    'vacation_season',
    'peak_commute']

target_A = 'y_arrivals_next_DT'
target_D = 'y_departures_next_DT'
```

```python
# 1. Compute demand lags and demand target
print("Computing demand lags and demand target...")

# Compute demand lags: dem_lag_i = arr_lag_i - dep_lag_i
demand_lag_cols = []
for i in range(1, 7):  # lags 1-6
    dem_lag_col = f'dem_lag_{i}'
    arr_lag_col = f'arr_lag_{i}'
    dep_lag_col = f'dep_lag_{i}'
    
    if arr_lag_col in df_merged.columns and dep_lag_col in df_merged.columns:
        df_merged = df_merged.with_columns(
            (pl.col(arr_lag_col) - pl.col(dep_lag_col)).alias(dem_lag_col)
        )
        demand_lag_cols.append(dem_lag_col)

# Compute demand target: target = y_arr - y_dep
df_merged = df_merged.with_columns(
    (pl.col(target_A) - pl.col(target_D)).alias('demand_target')
)

print(f"✅ Created demand lags: {demand_lag_cols}")
print(f"✅ Created demand target: demand_target")
```

```python
# 2. Update feature columns to include demand lags
feature_cols_demand = [
    'dep_last_DT',
    'trip_dur_mean_last_DT',
    'dep_lag_1', 'dep_lag_2', 'dep_lag_3', 'dep_lag_4', 'dep_lag_5', 'dep_lag_6',
    'arr_last_DT',
    'arr_lag_1', 'arr_lag_2', 'arr_lag_3', 'arr_lag_4', 'arr_lag_5', 'arr_lag_6',
    # Add demand lags
    'dem_lag_1', 'dem_lag_2', 'dem_lag_3', 'dem_lag_4', 'dem_lag_5', 'dem_lag_6',
    # Weather features
    'weather_temperature_2m',
    'weather_relative_humidity_2m',
    'weather_apparent_temperature',
    'weather_precipitation',
    'weather_wind_speed_10m',
    'weather_is_day',
    # Temporal features
    'is_holiday_ar',
    'sin_hour', 'cos_hour',
    'sin_dow', 'cos_dow',
    'sin_month', 'cos_month',
    'is_weekend',
    'payday_flag',
    'vacation_season',
    'peak_commute'
]

# Filter to only include features that exist in the dataframe
feature_cols_demand = [col for col in feature_cols_demand if col in df_merged.columns]

print(f"✅ Feature columns for demand prediction: {len(feature_cols_demand)} features")
print("Features:", feature_cols_demand)
```

```python
# 3. Prepare data for training
print("Preparing data for training...")

# Convert to pandas for sklearn compatibility
df_train = df_merged.to_pandas()

# Remove rows with null values in target or features
target_col = 'demand_target'
all_cols = feature_cols_demand + [target_col, target_A, target_D, 'cluster', 'ts_start']
df_train = df_train[all_cols].dropna()

print(f"Training data shape: {df_train.shape}")
print(f"Number of unique clusters: {df_train['cluster'].nunique()}")
print(f"Date range: {df_train['ts_start'].min()} to {df_train['ts_start'].max()}")

# Check target distributions
print(f"\nTarget statistics:")
print(f"Demand target ({target_col}):")
print(df_train[target_col].describe())
print(f"\nArrivals target ({target_A}):")
print(df_train[target_A].describe())
print(f"\nDepartures target ({target_D}):")
print(df_train[target_D].describe())

# Verify arrivals and departures are non-negative (as required for Poisson)
min_arrivals = df_train[target_A].min()
min_departures = df_train[target_D].min()

if min_arrivals < 0 or min_departures < 0:
    print(f"⚠️  Warning: Found negative values in count data!")
    print(f"   Min arrivals: {min_arrivals}, Min departures: {min_departures}")
else:
    print(f"✅ Arrivals and departures are non-negative (suitable for Poisson modeling)")
    print(f"   Min arrivals: {min_arrivals}, Min departures: {min_departures}")

print(f"✅ Data preparation complete")
```

```python
# Import additional libraries for modeling
from sklearn.linear_model import PoissonRegressor
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
import numpy as np

print("✅ Imported modeling libraries")
```

```python
# 5. Create Wide Feature Matrix (ALL clusters per timestamp)
print("Creating wide feature matrix...")
print("This will create one row per timestamp with features from ALL clusters")

# Import the new utilities
from cluster_utils import create_wide_feature_matrix, build_wide_cluster_dataset

# Create wide feature matrix: each feature duplicated for every cluster
# Example: dep_last_DT becomes dep_last_DT_0, dep_last_DT_1, ..., dep_last_DT_92
wide_X = create_wide_feature_matrix(df_merged, feature_cols_demand)

print(f"Wide feature matrix shape: {wide_X.shape}")
print(f"Number of unique timestamps: {len(wide_X.index)}")
print(f"Sample column names: {list(wide_X.columns[:10])}")
print("✅ Wide feature matrix created")
```

```python
# 6. Train Wide Models Per Cluster
print("=" * 60)
print("🎯 TRAINING WIDE MODELS PER CLUSTER")
print("=" * 60)
print("Each model will be trained on ALL cluster features but predict only ONE cluster's targets")

# Get unique clusters
unique_clusters = sorted(df_merged['cluster'].unique())
print(f"Training models for {len(unique_clusters)} clusters...")

# Store results for each cluster
cluster_results = []

# Loop over each cluster to train separate models
for i, cluster_id in enumerate(unique_clusters):
    print(f"\n[{i+1}/{len(unique_clusters)}] Training models for cluster {cluster_id}...")
    
    # Build dataset for this cluster (wide features + cluster-specific targets)
    try:
        df_cluster = build_wide_cluster_dataset(
            wide_X, df_merged, cluster_id,
            target_A='y_arrivals_next_DT',
            target_D='y_departures_next_DT'
        )
        
        print(f"  Dataset shape: {df_cluster.shape}")
        print(f"  Features: {df_cluster.shape[1] - 2} (all clusters)")
        print(f"  Samples: {len(df_cluster):,}")
        
        # Prepare features (all columns except targets)
        feature_cols_wide = [col for col in df_cluster.columns 
                           if col not in ['y_arrivals_next_DT', 'y_departures_next_DT']]
        
        # Train using the existing baseline function
        result = train_cluster_baseline_v2(
            cluster_data=df_cluster,
            cluster_id=cluster_id,
            feature_cols=feature_cols_wide,
            target_A='y_arrivals_next_DT',
            target_D='y_departures_next_DT',
            n_splits=3,
            poisson_max_iter=100000
        )
        
        cluster_results.append(result)
        
        if result['convergence_success']:
            print(f"  ✅ MAE: {result['mae_mean']:.4f}, RMSE: {result['rmse_mean']:.4f}, R²: {result.get('r2_mean', 'N/A'):.4f}")
        else:
            print(f"  ❌ Failed: {result.get('error', 'Unknown error')}")
            
    except Exception as e:
        print(f"  ❌ Error processing cluster {cluster_id}: {str(e)}")
        cluster_results.append({
            'cluster_id': cluster_id,
            'convergence_success': False,
            'error': str(e)
        })

print(f"\n✅ Completed training for {len(unique_clusters)} clusters")
successful_clusters = [r for r in cluster_results if r.get('convergence_success', False)]
print(f"Successful clusters: {len(successful_clusters)}/{len(unique_clusters)}")
```
