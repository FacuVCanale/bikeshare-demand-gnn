```python
import matplotlib.pyplot as plt
```

```python
from plotly.offline import init_notebook_mode
import plotly.graph_objects as go

# 1) Inicializar Plotly en modo offline para Jupyter
init_notebook_mode(connected=True)

# Datos
modelos = [
    "Poi. Reg.",
    "LGBM L=1",
    "LGBM L=6",
    "LGBM Hist.",
    "GTN"
]
r2 = [-0.071, 0.280, 0.340, 0.390, 0.462]

# Diccionario de mapeo de colores
color_map = {
    "Poi. Reg.": "#990b0b",
    "LGBM L=1": "#c41d1d",
    "LGBM L=6": "#c93131",
    "LGBM Hist.": "#f8d046",
    "GTN": "#3dda52"
}

# Generar lista de colores en el orden de los modelos
colores_personalizados = [color_map[modelo] for modelo in modelos]

# Opacidad deseada (alfa)
bar_opacity = 0.9  # valor entre 0 (transparente) y 1 (opaco)

# Crear figura con cada barra de un color específico y opacidad ajustable
fig = go.Figure(data=[
    go.Bar(
        x=modelos,
        y=r2,
        marker_color=colores_personalizados,
        opacity=bar_opacity
    )
])

# Configurar estética y dimensión (más ancha y menos alta)
fig.update_layout(
    template='plotly_white',
    width=1600,     # ancho en píxeles
    height=500,     # alto en píxeles
    xaxis_title='Modelo',
    xaxis_title_font=dict(size=24),
    yaxis_title='R\u00b2',
    yaxis_title_font=dict(size=24),
    xaxis_tickangle=-45,
    xaxis_tickfont=dict(size=24),
    yaxis_tickfont=dict(size=24),
    margin=dict(l=60, r=60, t=80, b=120),
    yaxis=dict(range=[min(r2) - 0.05, 0.5])
)

# Mostrar inline en el notebook
fig.show()

# Guardar localmente en alta resolución con misma proporción
fig.write_image(
    "comparacion_modelos_r2.png",
    width=700, 
    height=600, 
    scale=3
)
```
